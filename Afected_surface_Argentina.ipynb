```python
sup_total=df.superficie_afectada_por_incendios_total_hectareas.unique()
print(sup_total)
sup_total_numeric=pd.to_numeric(sup_total, errors='coerce')
print(sup_total_numeric)
```

```python
d=pd.DataFrame(sup_total,sup_total_numeric)
d
```

```python
#rename the dataframe
d.rename(columns={0:'string_value'}, inplace=True)
d.reset_index(inplace=True)
d.rename(columns={'index':'numeric_value'}, inplace=True)
print(d)

#we check if we have missing data
d.isna().sum()
```

```python
#we will evaluated the burned surface per year
burned_surface_per_year=[]
for i in years:
    y=df.loc[df['superficie_afectada_por_incendios_anio']==i].superficie_afectada_por_incendios_total_hectareas.sum().round(0)
    
    burned_surface_per_year.append(y)
```
